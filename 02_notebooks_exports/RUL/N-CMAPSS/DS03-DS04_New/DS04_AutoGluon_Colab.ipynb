```python
import json
import os
import random
import time
from pathlib import Path
from typing import Any, Dict, List, Optional, Tuple

import numpy as np
import pandas as pd
from autogluon.tabular import TabularPredictor

DATASET = 'DS04'
INPUT_ROOT = Path('/content/INPUT')
OUTPUT_ROOT = Path('/content/OUTPUT')
PRESET = 'high_quality'
SEED = 42
USE_CYCLE = True
FE_MODE = 'light'  # 'none' veya 'light'
USE_SCALER_FEATURES = True
VAL_ENGINES = 3
TIME_LIMIT = 3600

TARGET_COL = 'RUL'
ENGINE_COL = 'engine_id'
CYCLE_COL = 'cycle'
META_COLS = {'dataset_id', 'split'}

def set_seed(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

def json_default(obj: Any) -> Any:
    if isinstance(obj, (np.integer, np.floating)):
        return obj.item()
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    if isinstance(obj, Path):
        return str(obj)
    if hasattr(obj, 'to_dict') and callable(getattr(obj, 'to_dict')):
        return obj.to_dict()
    return str(obj)

def dump_json(path: Path, payload: Dict[str, Any]) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open('w', encoding='utf-8') as fp:
        json.dump(payload, fp, indent=2, default=json_default)

def validate_split_schema(df: pd.DataFrame, split_name: str, require_target: bool = True) -> None:
    required_cols = {ENGINE_COL, CYCLE_COL}
    if require_target:
        required_cols.add(TARGET_COL)
    missing = required_cols.difference(df.columns)
    if missing:
        raise ValueError(f'{split_name} missing required columns: {sorted(missing)}')
    if df[[ENGINE_COL, CYCLE_COL]].isna().any().any():
        raise ValueError(f'{split_name} contains null values in join keys')
    dup_count = int(df.duplicated(subset=[ENGINE_COL, CYCLE_COL]).sum())
    if dup_count > 0:
        raise ValueError(f'{split_name} has duplicated (engine_id, cycle) rows: {dup_count}')

def assert_no_engine_overlap(train_df: pd.DataFrame, test_df: pd.DataFrame) -> None:
    train_engines = set(train_df[ENGINE_COL].astype(str).unique())
    test_engines = set(test_df[ENGINE_COL].astype(str).unique())
    overlap = train_engines.intersection(test_engines)
    if overlap:
        sample = sorted(list(overlap))[:10]
        raise AssertionError(f'Engine leakage detected. overlap={len(overlap)}, sample={sample}')

def load_scaler_feature_columns(scaler_path: Path) -> Optional[List[str]]:
    if not scaler_path.exists():
        return None
    try:
        with scaler_path.open('r', encoding='utf-8') as fp:
            payload = json.load(fp)
    except Exception as exc:
        print(f'[WARN] scaler okunamadi: {exc}')
        return None
    cols = payload.get('feature_columns')
    if not isinstance(cols, list) or not cols:
        return None
    return [str(c) for c in cols]

def split_train_for_tuning(train_df: pd.DataFrame, val_engines: int, seed: int) -> Tuple[pd.DataFrame, Optional[pd.DataFrame], int]:
    if val_engines <= 0:
        return train_df.copy(), None, 0
    engines = np.asarray(sorted(train_df[ENGINE_COL].astype(str).unique()))
    if len(engines) < 2:
        return train_df.copy(), None, 0
    n_val = min(val_engines, len(engines) - 1)
    rng = np.random.RandomState(seed)
    val_set = set(rng.choice(engines, size=n_val, replace=False))
    val_df = train_df[train_df[ENGINE_COL].astype(str).isin(val_set)].copy()
    train_part = train_df[~train_df[ENGINE_COL].astype(str).isin(val_set)].copy()
    if val_df.empty or train_part.empty:
        return train_df.copy(), None, 0
    return train_part, val_df, n_val

def build_base_feature_cols(train_df: pd.DataFrame, use_cycle: bool, preferred_feature_cols: Optional[List[str]] = None) -> List[str]:
    excluded = {TARGET_COL, ENGINE_COL}.union(META_COLS)
    if preferred_feature_cols:
        candidate = [c for c in preferred_feature_cols if c in train_df.columns]
    else:
        candidate = [c for c in train_df.columns if c not in excluded]
    candidate = [c for c in candidate if c not in excluded]
    feature_cols = [c for c in candidate if pd.api.types.is_numeric_dtype(train_df[c])]
    feature_cols = list(dict.fromkeys(feature_cols))
    # FIX: scaler listesinde cycle olmasa da use_cycle=True ise ekle
    if use_cycle and CYCLE_COL in train_df.columns and CYCLE_COL not in feature_cols:
        feature_cols = [CYCLE_COL] + feature_cols
    if (not use_cycle) and (CYCLE_COL in feature_cols):
        feature_cols.remove(CYCLE_COL)
    if any(c.lower() == 'rul' or c.lower().startswith('rul_') for c in feature_cols):
        raise AssertionError('Potential target leakage: RUL-like column detected in features')
    if not feature_cols:
        raise ValueError('No usable numeric features found')
    return feature_cols

def build_fe_plan(train_df: pd.DataFrame, feature_cols: List[str], fe_mode: str) -> Dict[str, Any]:
    plan: Dict[str, Any] = {'fe_mode': fe_mode}
    if fe_mode == 'none':
        plan['base_numeric_cols'] = []
        return plan
    numeric_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(train_df[c]) and c != CYCLE_COL]
    plan.update({'base_numeric_cols': numeric_cols, 'rolling_window': 3, 'slope_window': 5})
    return plan

def apply_light_fe(df: pd.DataFrame, fe_plan: Dict[str, Any]) -> pd.DataFrame:
    if fe_plan['fe_mode'] == 'none':
        return df.copy()
    base_cols = fe_plan['base_numeric_cols']
    roll_w = int(fe_plan['rolling_window'])
    slope_w = int(fe_plan['slope_window'])
    out = df.sort_values([ENGINE_COL, CYCLE_COL]).copy()
    grp = out.groupby(ENGINE_COL, sort=False)
    for col in base_cols:
        out[f'{col}__delta1'] = grp[col].diff().fillna(0.0)
        out[f'{col}__roll_mean{roll_w}'] = grp[col].transform(lambda s: s.rolling(roll_w, min_periods=1).mean())
        out[f'{col}__roll_std{roll_w}'] = grp[col].transform(lambda s: s.rolling(roll_w, min_periods=1).std().fillna(0.0))
        out[f'{col}__slope{slope_w}'] = grp[col].transform(lambda s: s.diff().rolling(slope_w, min_periods=1).mean().fillna(0.0))
    return out

def build_final_feature_cols(base_feature_cols: List[str], train_fe: pd.DataFrame, fe_plan: Dict[str, Any]) -> List[str]:
    final_cols = list(base_feature_cols)
    if fe_plan['fe_mode'] == 'none':
        return final_cols
    roll_w = int(fe_plan['rolling_window'])
    slope_w = int(fe_plan['slope_window'])
    for col in fe_plan['base_numeric_cols']:
        eng = [f'{col}__delta1', f'{col}__roll_mean{roll_w}', f'{col}__roll_std{roll_w}', f'{col}__slope{slope_w}']
        for ec in eng:
            if ec in train_fe.columns:
                final_cols.append(ec)
    return list(dict.fromkeys(final_cols))

def rmse(y_true: pd.Series, y_pred: pd.Series) -> float:
    err = np.asarray(y_true) - np.asarray(y_pred)
    return float(np.sqrt(np.mean(np.square(err))))

set_seed(SEED)
OUTPUT_ROOT.mkdir(parents=True, exist_ok=True)
dataset_dir = INPUT_ROOT / DATASET
train_path = dataset_dir / f'train_{DATASET}_v0.csv'
test_path = dataset_dir / f'test_{DATASET}_v0.csv'
scaler_path = dataset_dir / f'scaler_{DATASET}_v0.json'

if not train_path.exists():
    raise FileNotFoundError(f'Train file not found: {train_path}')
if not test_path.exists():
    raise FileNotFoundError(f'Test file not found: {test_path}')

run_dir = OUTPUT_ROOT / f'ncmapss_{DATASET}_autogluon_artifacts'
run_dir.mkdir(parents=True, exist_ok=True)

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
validate_split_schema(train_df, 'train', require_target=True)
validate_split_schema(test_df, 'test', require_target=(TARGET_COL in test_df.columns))
assert_no_engine_overlap(train_df, test_df)

scaler_feature_cols = load_scaler_feature_columns(scaler_path) if USE_SCALER_FEATURES else None
train_part_raw, val_part_raw, val_engines_used = split_train_for_tuning(train_df, VAL_ENGINES, SEED)
base_feature_cols = build_base_feature_cols(train_part_raw, USE_CYCLE, scaler_feature_cols)

fe_plan = build_fe_plan(train_part_raw, base_feature_cols, FE_MODE)
train_fe = apply_light_fe(train_df, fe_plan)
train_part_fe = apply_light_fe(train_part_raw, fe_plan)
val_part_fe = apply_light_fe(val_part_raw, fe_plan) if val_part_raw is not None else None
test_fe = apply_light_fe(test_df, fe_plan)
final_feature_cols = build_final_feature_cols(base_feature_cols, train_fe, fe_plan)
if any(c.lower() == 'rul' or c.lower().startswith('rul_') for c in final_feature_cols):
    raise AssertionError('Potential target leakage in final feature columns')

train_model_df = train_part_fe[final_feature_cols + [TARGET_COL]].copy()
val_model_df = val_part_fe[final_feature_cols + [TARGET_COL]].copy() if val_part_fe is not None else None
has_test_rul = TARGET_COL in test_fe.columns
test_model_df = test_fe[final_feature_cols + [TARGET_COL]].copy() if has_test_rul else test_fe[final_feature_cols].copy()

predictor = TabularPredictor(label=TARGET_COL, eval_metric='rmse', path=str(run_dir / 'autogluon_models'))
fit_kwargs: Dict[str, Any] = {'train_data': train_model_df, 'presets': PRESET}
if val_model_df is not None:
    fit_kwargs['tuning_data'] = val_model_df
    fit_kwargs['use_bag_holdout'] = True
if TIME_LIMIT is not None:
    fit_kwargs['time_limit'] = TIME_LIMIT

started_at = time.time()
predictor.fit(**fit_kwargs)
elapsed_sec = time.time() - started_at

leaderboard_data = val_model_df if val_model_df is not None else (test_model_df if has_test_rul else train_model_df)
leaderboard = predictor.leaderboard(data=leaderboard_data, silent=True)
leaderboard_path = run_dir / 'leaderboard.csv'
leaderboard.to_csv(leaderboard_path, index=False)

train_part_pred = predictor.predict(train_model_df.drop(columns=[TARGET_COL]))
val_pred = predictor.predict(val_model_df.drop(columns=[TARGET_COL])) if val_model_df is not None else None
test_pred = predictor.predict(test_model_df.drop(columns=[TARGET_COL], errors='ignore'))

parts = [pd.DataFrame({'dataset_id': DATASET, 'split': 'train', ENGINE_COL: train_part_fe[ENGINE_COL].values, CYCLE_COL: train_part_fe[CYCLE_COL].values, 'rul_pred': np.asarray(train_part_pred)})]
if val_part_fe is not None and val_pred is not None:
    parts.append(pd.DataFrame({'dataset_id': DATASET, 'split': 'val', ENGINE_COL: val_part_fe[ENGINE_COL].values, CYCLE_COL: val_part_fe[CYCLE_COL].values, 'rul_pred': np.asarray(val_pred)}))
parts.append(pd.DataFrame({'dataset_id': DATASET, 'split': 'test', ENGINE_COL: test_fe[ENGINE_COL].values, CYCLE_COL: test_fe[CYCLE_COL].values, 'rul_pred': np.asarray(test_pred)}))
predictions_df = pd.concat(parts, ignore_index=True)
predictions_df['rul_pred'] = predictions_df['rul_pred'].clip(lower=0.0)

pred_path = OUTPUT_ROOT / f'ncmapss_{DATASET}_rul_predictions_autogluon.csv'
predictions_df.to_csv(pred_path, index=False)

model_info = {
    'dataset': DATASET,
    'best_model': predictor.model_best,
    'train_rows': int(len(train_model_df)),
    'test_rows': int(len(test_model_df)),
    'train_total_rows': int(len(train_fe)),
    'val_rows': int(len(val_model_df)) if val_model_df is not None else 0,
    'val_engines_used': int(val_engines_used),
    'n_features': int(len(final_feature_cols)),
    'train_part_rmse': rmse(train_model_df[TARGET_COL], train_part_pred),
    'val_rmse': rmse(val_model_df[TARGET_COL], val_pred) if val_model_df is not None else None,
    'test_rmse': rmse(test_fe[TARGET_COL], test_pred) if has_test_rul else None,
    'fit_elapsed_sec': elapsed_sec,
    'predictor_info': predictor.info(),
}
dump_json(run_dir / 'model_info.json', model_info)

run_config = {
    'dataset': DATASET,
    'seed': SEED,
    'preset': PRESET,
    'use_cycle': bool(USE_CYCLE),
    'fe_mode': FE_MODE,
    'use_scaler_features': bool(USE_SCALER_FEATURES),
    'val_engines': int(VAL_ENGINES),
    'val_engines_used': int(val_engines_used),
    'time_limit': TIME_LIMIT,
    'input_root': str(INPUT_ROOT.resolve()),
    'output_root': str(OUTPUT_ROOT.resolve()),
    'train_path': str(train_path.resolve()),
    'test_path': str(test_path.resolve()),
    'scaler_path': str(scaler_path.resolve()),
    'scaler_features_used': bool(scaler_feature_cols),
    'prediction_path': str(pred_path.resolve()),
    'artifacts_dir': str(run_dir.resolve()),
    'feature_count': int(len(final_feature_cols)),
    'feature_sample': final_feature_cols[:20],
}
dump_json(run_dir / 'run_config.json', run_config)

print('[OK] Predictions:', pred_path)
print('[OK] Leaderboard:', leaderboard_path)
print('[OK] Model info:', run_dir / 'model_info.json')
print('[OK] Run config:', run_dir / 'run_config.json')

if 'RUL' in test_df.columns:
    merged_test = test_df[['engine_id', 'cycle', 'RUL']].merge(
        predictions_df[predictions_df['split'] == 'test'][['engine_id', 'cycle', 'rul_pred']],
        on=['engine_id', 'cycle'],
        how='inner',
    )
    test_rmse = np.sqrt(np.mean((merged_test['rul_pred'] - merged_test['RUL']) ** 2))
    test_mae = np.mean(np.abs(merged_test['rul_pred'] - merged_test['RUL']))
    test_bias = np.mean(merged_test['rul_pred'] - merged_test['RUL'])
    print(f'Test RMSE: {test_rmse:.4f}')
    print(f'Test MAE : {test_mae:.4f}')
    print(f'Test Bias: {test_bias:+.4f}')

!cd /content && zip -r OUTPUT_DS04_autogluon.zip OUTPUT > /dev/null
print('Zip hazir: /content/OUTPUT_DS04_autogluon.zip')
```
